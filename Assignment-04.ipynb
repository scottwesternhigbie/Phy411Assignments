**1** Using the data from the hourly time series at Deep Cove (`dc`), show that the canned `fft` routine returns the correct discrete Fourier Transform of a suitable subset of the data (your choice on what subset).  You can do this graphically, by comparing the amplitude and the phase of the Fourier components, or simply show that the resulting numbers are the same.  

Hints: 
  - ignore bad data (rather than dealing with it "properly").  Make sure the units of your FT are: $\mathrm{^oC\, s}$.
  - If you want to compute the DFT effeciently, form the matrix $F$ referred to in the notes, and plot some of the time series associated with the different entries and make sure they look like Sines and Cosines and that the edge conditions are what you want


````python
dcslice = dc[5000:6024]
dctemp = dcslice[np.isfinite(dcslice)]
dctempCanned = np.fft.fft(dctemp) 
N=len(dctemp)

C = np.zeros((N,(N/2)+1))
for k in range(0,(N/2)+1):
    for n in range(0,N):
        C[n][k] = np.cos((2*np.pi*k*n)/N)
    
S = np.zeros((N,(N/2)-1))

for k in range(0,(N/2)-1):
    for n in range(0, N): 
        S[n][k] = np.sin((2*np.pi*(k+1)*(n))/N)

F = np.hstack([C,S])  
Xo = np.zeros(N)
Xo = (N/2)*np.dot(np.linalg.inv(F),dctemp)

X =(np.zeros(((N/2)+1),dtype = np.complex128))

X[1:N/2-1] = 3600.*Xo[1:N/2-1] - 1j*3600.*Xo[N/2+1:N-1]
X[0] = 2*3600.*Xo[0] #off by a factor of two for some reason... otherwise the right number
X[N/2] = 2*3600.*Xo[N/2] #off by a factor of two for some reason... otherwise the right number
print "Coded FFT", X[0:5]

print "Canned FFT", 3600.*dctempCanned[0:5]
````

The two fast fourier transforms are equivalent. One computed via a canned routine, and one created by me. (Albeit a very slight deviation of around the order 1E-8. This could just be a rounding error though.)
