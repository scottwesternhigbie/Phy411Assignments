## Q1: Distribution of power spectral estimates

<div style='background:#F0F0F0'>**1** You **may** use `matplotlib.mlab.psd` for the following question, but you had best use it correctly!

Using normally distributed random noise time series of length $N=2048$, show using a Monte Carlo analysis and by comparison to the theoretical probability distribution functions that:  
</div>

   1. The raw spectral estimate is indeed distributed as $\chi^2_2$.
   2. Show that block averaging with no overlap is distributed as $\chi^2_{2N_{blocks}}$.
   3. Show that block averaging with 50% overlap Hanning windows is distributed as $\chi^2_{2N_{blocks}}$.

<div style='background:#F0F0F0'>You can choose your block length, but making it an integer divisor of 2048 will make your life a lot easier.  Show that as $N_{blocks}$ is increased the variance drops (i.e. the distribution of the spectral estimates gets tighter).</div>


<div style='background:#F0F0F0'>HINT: for this time series the individual frequency estimates are indipendent samples of the distribution, so you can use them in compiling your distributions.</div>

<div style='background:#F0F0F0'>HINT: To get the pdf of the $\chi^2_\nu$ distribution correct, you need to "scale" by $\nu^{-1}$ where $\nu$ are the degrees of freedom.</div>

<div style='background:#F0F0F0'>HINT: for good presentation, make your histograms have the same bin sizes, and compare the distributions for all three cases on the same plot.</div>

```python
fig1 = plt.figure(figsize=(12,8))
ax = fig1.add_subplot(1,1,1)

#Raw Spectral Estimate
N = 2048.
T = N
t = np.arange(N)
dt = T/N 
f = np.arange(N/2)/T
Gxs = np.zeros((100,N/2))
for i in range(100):
    x = (np.random.randn(N))
    X = dt*np.fft.fft(x)
    Gxx = 2.*np.real(np.conj(X[:N/2])*X[:N/2])/T
    Gxs[i,:] = Gxx
    f = np.arange(N/2)/T
    ax.loglog(f,Gxs[i],".", markersize = 1, color = 'b')
    
##Welch's Block Averaging Estimate (No overlap)   
#NFFT=256
#nblock = np.floor(N/NFFT)
#Gblock = np.zeros((nblock, NFFT/2))
#for i in range(100):
    #xb=np.random.randn(N)
    #for j in range(int(nblock)):
        #js = range(j*NFFT,(j+1)*NFFT)
        #tblock = t[js]
        #xblock = xb[js]
        #Xb = dt*np.fft.fft(xblock)
        #Tblock=T/nblock
        #Gblock[j,:] = (2./Tblock)*np.real(np.conj(Xb[:NFFT/2])*Xb[:NFFT/2])
        #fblock= np.arange(NFFT/2)/Tblock
        ##ax.loglog(fblock[1:],Gblock[j,1:],'.',color='r')
        #Gblockm = np.mean(Gblock, axis = 0)
        #ax.loglog(fblock,Gblockm,'.', markersize = 1, color = 'r')
        
#Canned psd method (No window)
GCs = np.zeros((100,129))
for i in range(100):    
    x = np.random.randn(N)
    GCanned, fCanned = mlab.psd(x, Fs = 1., window=mlab.window_none)
    ax.loglog(fCanned,GCanned, '.', markersize = 1, color = 'k')       
    GCs[i,:] = GCanned    

#Canned psd method (50% Hanning window)
GCHs = np.zeros((100,129))
for i in range(100):
    x = np.random.randn(N)
    GCannedH, fCannedH = mlab.psd(x,Fs = 1., window = mlab.window_hanning, noverlap = 128)
    ax.loglog(fCannedH,GCannedH, '.', markersize = 1, color = 'm')
    GCHs[i,:] = GCannedH
    
ax.set_xlabel("$f [Hz]$")
ax.set_ylabel("$G_{xs}(f_k)$")
ax.set_title("Monte Carlo Simulation with Spectral Estimates")
fig1.tight_layout()
```

Blue is the raw spectral estimate, black is the block averaging with no window, and magenta is the hanning window estimate. 

```python
#Expected Chi-Square Distributions
tl = np.linspace(0.,100., 100)
fig2 = plt.figure(figsize=(12,8))
ax = fig2.add_subplot(1,1,1)
for v in [2,2*(2048/256)]:
    y=stats.chi2.pdf(tl,v)
    ax.plot(tl,y,label=('Theoretical Chi Square Distributions: v=%2d'%v))

#Spectral Estimates Distributions
ChiR = Gxs/np.mean(Gxs, axis = 0)
ChiB = GCs/np.mean(GCs, axis = 0)
ChiH = GCHs/np.mean(GCHs, axis = 0)
#print np.shape(ChiR)
#print np.shape(ChiB)
#print np.shape(ChiH)
ChiR = np.reshape(ChiR, -1)
ChiB = np.reshape(ChiB, -1)
ChiH = np.reshape(ChiH, -1)
#print ChiR
#print ChiB
#print ChiH
ax.hist(ChiR*2.,bins=128,normed=True,color='m', histtype='step',alpha=0.5,linewidth = 3, label = 'Raw Chi')
ax.hist(ChiB*16.,bins=128,normed=True,color='k', histtype='step',alpha=0.5, label = 'No window, block chi')
ax.hist(ChiH*16.,bins=128,normed=True,color='b', histtype='step',alpha=0.5, label = 'Hanning Chi')
ax.legend();ax.set_title(r'$\chi^2_v$ distributions'); ax.set_xlabel('$G_{xx}$');ax.set_ylabel('PDF')
```

The Raw Chi Spectra histogram appears to follow the same distribution as the theoretical chi squared distribution, as does the Blocks average with no overlap. However, the hanning chi histogram does not follow the theoretical distribution for 2Nblocks = $v$ (To be expected however, as $v$ for the hanning window is 18*N/11).

```python
#Canned psd method (No window)
GCs2 = np.zeros((100,65))
for i in range(100):    
    x = np.random.randn(N)
    GCanned, fCanned = mlab.psd(x,NFFT = 128, Fs = 1., window=mlab.window_none)      
    GCs2[i,:] = GCanned    

#Canned psd method (50% Hanning window)
GCHs2 = np.zeros((100,65))
for i in range(100):
    x = np.random.randn(N)
    GCannedH, fCannedH = mlab.psd(x,NFFT = 128, Fs = 1., window = mlab.window_hanning, noverlap = 64)
    GCHs2[i,:] = GCannedH

#Expected Chi-Square Distributions
tl = np.linspace(0.,100., 100)
fig2 = plt.figure(figsize=(12,8))
ax = fig2.add_subplot(1,1,1)
for v in [2*(2048/128)]:
    y=stats.chi2.pdf(tl,v)
    ax.plot(tl,y,label=('Theoretical Chi Distribution for: v=%2d'%v))

#Spectral Estimates Distributions
ChiB2 = GCs2/np.mean(GCs2, axis = 0)
ChiH2 = GCHs2/np.mean(GCHs2, axis = 0)
ChiB2 = np.reshape(ChiB2,-1)
ChiH2 = np.reshape(ChiH2,-1)
ax.hist(ChiB2*32.,bins=128,normed=True,color='k', histtype='step',alpha=0.5,label ='(Nblocks = 16) No window, block chi')
ax.hist(ChiH2*32.,bins=128,normed=True,color='b', histtype='step',alpha=0.5, label = '(Nblocks = 16) Hanning Chi')
ax.hist(ChiB*16.,bins=128,normed=True,color='g', histtype='step',alpha=0.5, label = '(Nblocks = 8) No window, block chi')
ax.hist(ChiH*16.,bins=128,normed=True,color='r', histtype='step',alpha=0.5, label = '(Nblocks = 8) Hanning Chi')
ax.legend();ax.set_title(r'$\chi^2_v$ distributions'); ax.set_xlabel('$G_{xx}$');
ax.set_ylabel('PDF')
```

As NBlocks increases, the variance increases. If you were to increase NFFT, you would decrease variance. 

<div style='background:#F0F0F0'>**2**  For the Hanning window estimate, what fits better $2N_{blocks}$ degrees of freedom or $18N_{blocks}/11$ degrees of freedom (the correct theoretical value)?  Its OK to evaluate by eye.</div>

```python
#Expected Chi-Square Distributions
tl = np.linspace(0.,100., 100)
fig2 = plt.figure(figsize=(12,8))
ax = fig2.add_subplot(1,1,1)
for v in [2,2*(2048/128),26]:
    y=stats.chi2.pdf(tl,v)
    ax.plot(tl,y,label=('Theoretical Chi Distribution for: v=%2d'%v))
    
ax.hist(ChiH*16.,bins=128,normed=True,color='r', histtype='step',alpha=0.5, label = '(Nblocks = 16) 2N-Hanning Chi')
ax.hist(ChiH*16.*18./11.,bins=128,normed=True,color='k', histtype='step',alpha=0.5, label = '(Nblocks = 16) 18N/11-Hanning Chi')
ax.legend();ax.set_title(r'$\chi^2_v$ distributions'); ax.set_xlabel('$G_{xx}$');
ax.set_ylabel('PDF')
```

By changing NBlocks to 18*N/11, the chi squared distribution for the hanning spectra coincides with the theoretical chi squared distribution for n = 18*Nblocks/11. Therefore, 18*N/11 fits better than 2N.

## Q2: Spectra of a "red-noise" signal

<div style='background:#F0F0F0'> **1** Load in the Deep Cove hourly data, and compute the power spectra using some reasonable value for $N_{FFT}$, and comment on the effect of applying the Hanning window to not applying it to the spectral leakage.</div>

```python
hourdata=np.genfromtxt('http://web.uvic.ca/~jklymak/Phy411/Data/AllHourly.txt')[[6,28],2:]
dc = hourdata[0,:]
dc=dc[np.isfinite(dc)]
print len(dc)
```

```python
#Due to the fact that the length of dc is not even, it is somewhat difficult to find 
#an appropriate NFFT. Since the length is close to 24000, I have chose to set NFFT
#to 600.

NW = dc
X = np.fft.fft(NW)
M = len(NW)
T = M
Gnw = (2./T)*np.real(np.conj(X[:M/2])*X[:M/2])
fig = plt.figure(figsize=(12,8))
ax = fig.add_subplot(1,1,1)
f = np.arange(M/2)/(M*1.)
ax.loglog(f,Gnw, color = 'c', linewidth = 0.5, label = 'Raw')

W = dc
Gcanned,fcanned = mlab.psd(W,NFFT=600,Fs=1.,window=mlab.window_hanning,noverlap=256)
ax.loglog(fcanned,Gcanned,label='Canned', color = 'r', linewidth = 0.5)

ax.set_xlabel("$f [Hz]$")
ax.set_ylabel("$G_{xs}(f_k)$")
ax.set_title("Hanning Window Spectra and Raw Power Spectra vs. Frequency")
fig.tight_layout()
```

The raw spectra appears noisy with multiple peaks. The expected trend is present as we have seen before.

The windowed data appears sharp and the peaks appear periodically. The windowed data is also on a smaller range than the dc spectra, with the peaks more noticeable than in the noisey raw spectrum.
