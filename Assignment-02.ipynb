## Q1: Show that the correlation co-efficient is indeed related to the fraction of varaince explained by a linear dependence.

**1** To start, make a $N(0,1)$ random variable $x$, with 200,000 entries.  Then make a linearly dependent variable $y_i= a x_i + n_i$, where a is some amplitude, and $n_i$ is a random noise varaible.  Assume that $n_i$ is distributed as $N(0,n)$, where $n$ is the noise level.

Calculate the correlation co-efficient $r_{xy}$ between $x$ and $y$ for 40 different values of $n$, logarithmically spaced between 0.01 and 100. (hint use `np.logspace`).  

Plot $r_{xy}$ versus the inverse of the noise $n$ normalized by the standard devaition of the signal $ax_i$: $\frac{a s_x}{n}$   You may want to use `ax.loglog` to visualize the orders of magnitude.  

Comment on the resulting comparison.
<hr>

```python
import numpy as np
import matplotlib.pyplot as plt


n = np.logspace(-2.0,2.0,40)
a = 1 
r = []*40
for i in range (0,40):
    x = np.random.randn(200000)
    x = x - np.mean(x)
    #print "Mean and Standard Deviation for x: ", np.mean(x), np.std(x)
    y = a*x + np.random.randn(200000)*n[i]
    y = y - np.mean(y)
    #print "Mean and Standard Deviation for y: ", np.mean(y), np.std(y)
    r.append((np.sum(x*y))/((np.std(x,ddof=1)*np.std(y,ddof=1)*(200000-1))))
    #print "Correlation is: "
print "Correlation values r for 40 different values of n: ", r


plt.figure(figsize=(20,20))
plt.subplot(2,1,1)
plt.plot((a*np.std(x))/n,r)
plt.axis([0, 10, 0, 1.05])
plt.xlabel("Inverse Noise (1/n)", fontsize = 16)
plt.ylabel("Correlation (r)", fontsize = 16)
plt.legend(["Graph of Correlation vs. Inverse Noise"])


plt.subplot(2,1,2)
plt.loglog((a*np.std(x,ddof=1))/n,r)
plt.axis([-10, 100, -0.5, 1.5])
plt.xlabel("log of Inverse Noise log(1/n)", fontsize = 16)
plt.ylabel("Log of Correlation log(r)", fontsize = 16)
plt.legend(["Log-Log Plot of Correlation vs. Inverse Noise"])

```

The log-log plot appears to be more linear from ~n = 0.01 to 10 than the r vs inverse n plot. Both graphs tend to converge to correlation r = 1 as n becomes smaller, which should make sense, since x and y are essentially scalar multiples of each other when n = 1. I.e. take n = 1, and you have y = ax + N(0,1). Since x is N(0,1) anyway, y tends to y = (a+1)x.

To answer the question posed, the linear dependance of correlation can be visualized in the second plot above. Where the graph appears bent and "shaky", I believe that is where the noise has interfered with the linear relationship between x and y.

## Q2: Monte Carlo on slopes and test versus student-t distribution

**1** We stated without proof that the slope $b$ of a linear fit of $y$ to $x$ is distributed as a student-t distribution with $N-2$ degrees of freedom, with a standard devaition given by the standard deviation of $y$ around the fit model $\tilde{y}$.   Lets test that emperically using a Monte Carlo technique.  

First create an independet variable $x$ from 10 points from an $N(0,1)$ distribution.

Next, using $n=1.0$, and $a=1.0$ create the dependent variables $y_i=a x_i + n_i$ where $n_i$ are drawn from the $N(0,n)$ distribution.  Calculate the slope $b$.  

Do this 10000 times, and make a histogram of the resulting fits $b$.  Also histogram $s_{y|x}$ as defined in the notes.

Comment on the histigrams

<hr>

```python
b = []*10000
n = 1.0
a = 1.0
s = []*10000
for i in range(0,10000):
    x = np.random.randn(10)
    y = a*x + np.random.randn(10)*n
    c = (np.sum(x*y)-(np.mean(x)*np.mean(y)*10))/(np.sum(x**2)-(10*np.mean(x)*np.mean(x))) #intermediate variable
    b.append((np.sum(x*y)-(np.mean(x)*np.mean(y)*10))/(np.sum(x**2)-(10*np.mean(x)*np.mean(x))))
    yhat = c*x+a
    s.append(np.sqrt(np.sum(((y-yhat)**2))/8))
plt.figure(figsize=(20,20))
plt.subplot(2,2,1)
plt.hist(b, rwidth = 0.8)
plt.title("Slope, b", fontsize = 20)
plt.xlabel("Value", fontsize = 20)
plt.ylabel("Occurence", fontsize = 20)

plt.subplot(2,2,2)
plt.hist(s, rwidth = 0.8)
plt.title("Standard Deviation, $s_{y|x}$", fontsize = 20)
plt.xlabel("Value", fontsize = 20)
plt.ylabel("Occurence", fontsize = 20)
```

The slope histogram follows the expected trend of being very close to 1, since the amplitude variable $a$ is 1 and x is of normal distribution (between 0 and 1, with higher probability of being 1). The histogram also appears to follow the normal distribution trend. The standard deviation allows follows a normal trend, with the apparent most likely value being ~1.5 = Sigma.
