## Q1: Simulate aliasing of sinusoidal signals

**1** To start, make a time vector that goes from 0 to 100 s, with 10000 entries in it.  Then make a synthetic time series that has two close in frequency sine waves:

$$ x = A_1 \cos (\omega t) + A_2 \cos ((\omega+\delta \omega ) t) $$

where $\omega$ is the "carrier" frequency, and $\delta \omega << \omega$ is the difference in frequencies between the two waves.  For simplicty, you could make $A_1=A_2=1.0\ [V]$.  Feel free to choose an interesting $\omega$.

<hr>

```python
import numpy as np
time = np.arange(0,100,(100./10000.),float)
a = np.cos(0.5*time)
b = np.cos((0.5+0.05)*time)
x = np.cos(0.5*time)+ np.cos((0.5+0.05)*time)

```

## Q2: Demonstrate the central limit theorem for a very unfair coin flip.

Consider a coin that is twice as likely to come up heads as tails. To put a number on it, call tails $x=-1/2$ and heads $x=+1/2$.  

**1** What are the mean, $\mu_x$, and variance $\sigma_x^2$ of this process?

<hr>

The mean of the coin flip is 0.166... (which is simply the sum of the indivual products of the outcome and probability of each possiblity).

The variance of the coin flip is given by: variance = sigma((outcome - mean)^2 *probability of ith outcome). In this case, the variance was found to be 0.222...

```python
#from q = 1- p, if p is 2*q, then q must be 1/3 and p 2/3, unless you consider some crazy universe with >1 probability...
p1 = (2.0/3.0)
p2 = (1.0/3.0)
x1 = 0.5
x2 = -0.5
mean = (x1*p1+x2*p2)
print "The mean =", mean

variance = ((x1-mean)*(x1-mean)*p1) + ((x2-mean)*(x2-mean)*p2)
print "The variance =", variance
```

**2** Plot or represent the *probability distribution*  (or cumulative distrubution) of this process.
<hr>

Numpy exponent?


```python
z = (1/((np.pi*2*variance)**(1/2)))*np.exp(-((x-mean)**2)/(2*variance))
plt.figure(figsize=(9,6))
plt.plot(x,z)
plt.xlabel("Outcome", fontsize=18)
plt.ylabel("Probability", fontsize = 18)
plt.axis([-3,3,0,1])
plt.legend(["Probability Distribution for an unfair coin flip"])
```

**3** Write a function using the `np.random.rand` function to simulate the coin flip.  It should return $+1/2$ twice as often as $-1/2$.  Demonstrate that your function works.
<hr>

```python
def getFlip(N,printer = False):   
    aa = np.random.rand(N)-(1.0/2.0)
    bb = []*len(aa)
    for j in range(0,len(aa)):
        if(aa[j]>(-1.0/6.0)): bb.append((1.0/2.0)) 
        else: bb.append((-1.0/2.0))
    if printer:
        print 'Numer of flips: %d' % N
        print 'Mean: %1.3f'%np.mean(bb)
        print 'Variance: %1.3f'%np.var(bb)
        print ''
    return bb
    

N2000 = getFlip(10000,printer=True)

#You will observe that the mean and variance agree with my earlier calculations!
```

**4** Perform $N_{flip}$ and take the sample mean $\overline{x}$ of the values.  Do this 2000 times, and show that as $N_{flip}\to\infty$ the distribution of $\overline{x}$ approaches a Normal distribution by plotting the histograms of the means for each.  For $N_{flip}$ choose 2, 16, 64, 256. 

Comment on how the convergence of $\overline{x}$ for the unfair coin compares to the convergence of the sample mean of the dice roll.  
<hr>

```python
import scipy.stats as stats
plt.figure(figsize=(30,20))

Nexpts = 2000
means=np.zeros(Nexpts)
for i in range(Nexpts):
    xx=getFlip(2,printer = False)
    means[i]=np.mean(xx)
means2=means
print "For 2 = ", means2
plt.subplot(8,1,1)
plt.hist(means2)
plt.xlabel('Mean', fontsize = 18), plt.ylabel('Occurences', fontsize = 18)

Nexpts = 2000
means=np.zeros(Nexpts)
for i in range(Nexpts):
    xx=getFlip(16,printer = False)
    means[i]=np.mean(xx)
means16=means
print "For 16 =", means16
plt.subplot(8,1,3)
plt.hist(means16)
plt.xlabel('Mean', fontsize = 18), plt.ylabel('Occurences', fontsize = 18)

Nexpts = 2000
means=np.zeros(Nexpts)
for i in range(Nexpts):
    xx=getFlip(64,printer = False)
    means[i]=np.mean(xx)
means64=means
print "For 64 =", means64
plt.subplot(8,1,5)
plt.hist(means64)
plt.xlabel('Mean', fontsize = 18), plt.ylabel('Occurences', fontsize = 18)

Nexpts = 2000
means=np.zeros(Nexpts)
for i in range(Nexpts):
    xx=getFlip(256,printer = False)
    means[i]=np.mean(xx)
means256=means
print "For 256 =", means256
plt.subplot(8,1,7)
plt.hist(means256)
plt.xlabel('Mean', fontsize = 18), plt.ylabel('Occurences', fontsize = 18)
```

As $N_{flip}\to\infty$, we see the histograms proceed towards a normal distribution about the expected mean (from part 1). There appears to be negative skew however, but this is probably due to the fact that the number of flips is so low for each populations. Increase the flip count to infinity and you should be golden. 

In regards to the convergence of the biased coin flip and the unbiased dice roll, it appears that the dice roll function converges faster to normal than the coin flip, as the coin flip is inherently non-symmetrical, whereas the dice roll has equal probabilities for dice role outcome on either side of the mean, i.e. equal probability for 1/6 to 3/6 as there is for 4/6 to 6/6 relative to a middle mean of ~3.5/6 or so.

**5** Replot the histograms, removing the mean and normalizing by the standard deviation.  Also show the CDF for normal distribution with mean zero and standard deviation of 1.

```python
import scipy.stats as stats
plt.figure(figsize=(30,20))

plt.subplot(8,1,1)
plt.hist((means2-np.mean(means2))/np.std(means2))
plt.xlabel('Normalized Values', fontsize = 18), plt.ylabel('Occurences', fontsize = 18)

plt.subplot(8,1,3)
plt.hist((means16-np.mean(means16))/np.std(means16))
plt.xlabel('Normalized Values', fontsize = 18), plt.ylabel('Occurences', fontsize = 18)

plt.subplot(8,1,5)
plt.hist((means64-np.mean(means64))/np.std(means64))
plt.xlabel('Normalized Values', fontsize = 18), plt.ylabel('Occurences', fontsize = 18)

plt.subplot(8,1,7)
plt.hist((means256-np.mean(means256))/np.std(means256))
plt.xlabel('Normalized Values', fontsize = 18), plt.ylabel('Occurences', fontsize = 18)

x = np.arange(0,1)
plt.plot(x,stats.norm.cdf())
```
