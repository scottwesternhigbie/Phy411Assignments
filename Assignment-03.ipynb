## Q1: Lag correlations of weather time series

````python
hourdata=np.genfromtxt('http://web.uvic.ca/~jklymak/Phy411/Data/AllHourly.txt')[[6,28],2:]
dc = hourdata[0,:]
#jb = hourdata[1,:]
````

**1** From the hourly time series at Deep Cove (`dc`), calculate the *lag correlation co-efficient* $\rho_{xy}(\tau)$ out to a lag of $\tau=30\ \mathrm{days}$ (Umm, remember this data is hourly!).  (also rememeber to remove the mean of $dc$ before trying to to the calculations).  

Comment on the resulting plot.

````python
dcd = dc[np.isfinite(dc)]
Nt = 26280 #The data crunched in this question is for three years total

tau = np.zeros(31)
for i in range(0,31):
    tau[i] = i
    
dcp = np.zeros(Nt)
rho = np.zeros(31)

for taut in range(0,31):
    dcp = np.zeros(Nt)
    dcr = dc
    dcp = ((dc-np.mean(dcd))*(np.roll(dcr,taut*24)-np.mean(dcd)))
    dcp = dcp[np.isfinite(dcp)]
    rho[taut] = (np.mean(dcp)/(np.var(dcd)))
print "Lag Correlation Co-Efficient Array"
print rho

plt.figure(figsize=(10,10))
plt.xlabel("Lag Time in Days ($tau$)", fontsize = 15)
plt.ylabel("Lag Correlation Coefficient ($rho$)", fontsize = 15)
plt.title("Graph of Lag Correlation Coefficient vs. Lag in Days", fontsize = 15)
plt.plot(tau,rho)
````

One would expect that as the correlation coefficient becomes larger, the lag correlation would increase (as temperature data is a whole month ahead of the original data). This means that temperature data from say August and September would have terrible correlation, which is to be expected as the Fall is cold and summer is hot. If however, the lag correlation jumped up to something like 365 days, one would expect the correllation to be much closer to 1. Ex. even though the temperature day for Dec 31st (2013) will not be the exact same as Dec 31st (2014), one would expect to see a relatively close correlation due to December being a traditionally cold month (not counting global warming etc.). 

From the above paragraph, it seems that if one were to expand the lag time to a larger scale, you would observe an oscillating curve (possibly sinusoidal or cosine) from $\tau$ = 0 to $\tau$ = 365 days.

*2* Now, instead of doing the whole year, look at just the winter months and the summer months and compare the lag correlations (no need to do this for the whole time series, just choose one winter and one summer).  Make sure to indicate which lag correlation is which on the plots.

Comment below on the different charcter of these two lag correlations.

````python
dcw = dc[3:1491]

shift = 24*31*5
dcs = dc[3+shift:1491+shift]

tausw = np.zeros(31)
for i in range(0,31):
    tausw[i] = i
    
dcj = np.zeros(len(dcs))
rhow = np.zeros(31)
rhos = np.zeros(31)

#For Winter Months
for taut in range(0,31):
    dcj = np.zeros(len(dcs))
    dco = dcw
    #dcj = ((dcw-np.mean(dcw))*(np.roll(dco,taut*24)-np.mean(dcw)))
    dcj = ((dcw-np.mean(dcw))*(dc[3+(taut*24):1491+(taut*24)]-np.mean(dc[3+(taut*24):1491+(taut*24)])))
    dcj = dcj[np.isfinite(dcj)]
    rhow[taut] = (np.mean(dcj)/(np.var(dcw)))
print "Lag Correlation Coefficient Array (Winter)"
print rhow

#For Summer Months
for taut in range(0,31):
    dcj = np.zeros(len(dcs))
    dco = dcs
    #dcj = ((dcs-np.mean(dcs))*(np.roll(dco,taut*24)-np.mean(dcs)))
    dcj = ((dcs-np.mean(dcs))*(dc[3+shift+(taut*24):1491+shift+(taut*24)]-np.mean(dc[3+shift+(taut*24):1491+shift+(taut*24)])))
    dcj = dcj[np.isfinite(dcj)]
    rhos[taut] = (np.mean(dcj)/(np.var(dcs)))
print "Lag Correlation Coefficient Array (Summer)"
print rhos

plt.figure(figsize=(15,15))

plt.subplot(2,1,1)
plt.xlabel("Lag Time in Days ($tau$)", fontsize = 15)
plt.ylabel("Lag Correlation Coefficient ($rho$)", fontsize = 15)
plt.title("Graph of Lag Correlation Coefficient vs. Lag in Days (Winter Edition)", fontsize = 15)
plt.plot(tausw,rhow)

plt.subplot(2,1,2)
plt.xlabel("Lag Time in Days ($tau$)", fontsize = 15)
plt.ylabel("Lag Correlation Coefficient ($rho$)", fontsize = 15)
plt.title("Graph of Lag Correlation Coefficient vs. Lag in Days (Summer Edition)", fontsize = 15)
plt.plot(tausw,rhos)
````

In comparing the two plots, summer and winter, it appears that the summer months are more correlated to their lagged time series than the winter months. The winter months also produce a more oscillatory curve of $\rho_{xx}$.
